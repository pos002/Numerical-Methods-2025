# Лабораторная работа 5: Приближённое решение задачи Коши для ОДУ

**Вариант 26**

Уравнение: 
$$u'' + \frac{1}{\cos^2 x} u' + 2 \frac{\tg x}{\cos^2 x} u = 2 + 2x \frac{1 + x \tg x}{\cos^2 x}$$

Начальные условия: $u(0) = 1$, $u'(0) = -1$

Точное решение: $u_0(x) = x^2 + \exp(-\tg x)$

Шаг сетки: $h = 0.05$

```python
def exact_solution(x):
    """Точное решение"""
    return x**2 + np.exp(-np.tan(x))

def f(x, y):
    """
    Правая часть системы ОДУ
    y[0] = u, y[1] = u'
    """
    y1, y2 = y
    cos_x = np.cos(x)
    tg_x = np.tan(x)
    
    # Преобразуем уравнение к системе первого порядка
    dy1dx = y2  # u' = y2
    
    # u'' = - (1/cos²x)u' - 2(tgx/cos²x)u + 2 + 2x(1 + x tgx)/cos²x
    dy2dx = (-1/(cos_x**2) * y2 - 2*(tg_x/(cos_x**2)) * y1 + 
             2 + (2*x*(1 + x*tg_x))/(cos_x**2))
    
    return np.array([dy1dx, dy2dx])
```

```python
# Начальные условия
y0 = np.array([1.0, -1.0]) # y1(0) = u(0) = 1, y2(0) = u'(0) = -1

# Параметры сетки (ограничиваем x для избежания особенностей при x = π/2)
h = 0.05
x_max = 1.0  # Ограничиваем область до 1.0 радиуса для избежания особенностей
x_fine = np.arange(0, x_max + h, h)      # Подробная сетка
x_coarse = np.arange(0, x_max + 0.1, 0.1)  # Грубая сетка для оценки по Рунге
```

```python
def euler_method(f, y0, x):
    """Метод Эйлера"""
    n = len(x)
    y = np.zeros((n, len(y0)))
    y[0] = y0
    
    for i in range(n-1):
        h = x[i+1] - x[i]
        y[i + 1] = y[i] + h * f(x[i], y[i])
        
    return y
```

```python
def runge_kutta_4(f, y0, x):
    """Метод Рунге-Кутты 4-го порядка"""
    n = len(x)
    y = np.zeros((n, len(y0)))
    y[0] = y0
    
    for i in range(n-1):
        h = x[i + 1] - x[i]
        k1 = f(x[i], y[i])
        k2 = f(x[i] + h/2, y[i] + h*k1/2)
        k3 = f(x[i] + h/2, y[i] + h*k2/2)
        k4 = f(x[i] + h, y[i] + h*k3)
        
        y[i + 1] = y[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    
    return y
```

```python
def adams_3(f, y0, x_points):
    """Метод Адамса 3-го порядка"""
    n = len(x_points)
    y = np.zeros((n, len(y0)))
    y[0] = y0
    
    # Первые 3 точки находим методом Рунге-Кутты
    for i in range(2):
        h = x_points[i+1] - x_points[i]
        k1 = f(x_points[i], y[i])
        k2 = f(x_points[i] + h/2, y[i] + h*k1/2)
        k3 = f(x_points[i] + h/2, y[i] + h*k2/2)
        k4 = f(x_points[i] + h, y[i] + h*k3)
        y[i+1] = y[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    
    # Основная формула Адамса
    for i in range(2, n-1):
        h = x_points[i+1] - x_points[i]
        f_curr = f(x_points[i], y[i])
        f_prev = f(x_points[i-1], y[i-1])
        f_prev2 = f(x_points[i-2], y[i-2])
        
        y[i+1] = y[i] + (h/12) * (23*f_curr - 16*f_prev + 5*f_prev2)
    
    return y
```

```python
def calculate_errors(y_numerical, y_exact):
    """Вычисление погрешностей"""
    return np.abs(y_numerical - y_exact)

def runge_error_estimate(y_fine, y_coarse, p=4):
    """Оценка погрешности по Рунге"""
    return np.abs(y_fine - y_coarse) / (2**p - 1)
```

## Решение задачи

```python
# Вычисление решений
y_euler = euler_method(f, y0, x_fine)
y_rk4 = runge_kutta_4(f, y0, x_fine)
y_adams = adams_3(f, y0, x_fine)
y_exact = exact_solution(x_fine)

# Решение на грубой сетке для оценки по Рунге
y_rk4_coarse = runge_kutta_4(f, y0, x_coarse)
```

## Таблица результатов

```python
print("Сравнение численных методов:")
print("x\t\tЭйлер\t\tРунге-Кутта\tАдамс\t\tТочное")
print("-" * 70)
for i in range(len(x_fine)):
    if i % 2 == 0:  # Выводим каждую вторую точку для компактности
        print(f"{x_fine[i]:.2f}\t\t{y_euler[i,0]:.6f}\t{y_rk4[i,0]:.6f}\t{y_adams[i,0]:.6f}\t{y_exact[i]:.6f}")
```

## Погрешности методов

```python
print("\nПогрешности методов:")
print("x\t\tПогр. Эйлера\tПогр. Рунге-Кутта\tПогр. Адамса")
print("-" * 70)
for i in range(len(x_fine)):
    if i % 2 == 0:
        error_euler = abs(y_euler[i,0] - y_exact[i])
        error_rk4 = abs(y_rk4[i,0] - y_exact[i])
        error_adams = abs(y_adams[i,0] - y_exact[i])
        print(f"{x_fine[i]:.2f}\t\t{error_euler:.2e}\t\t{error_rk4:.2e}\t\t{error_adams:.2e}")
```

## Оценка точности по Рунге

```python
print("\nОценка точности по Рунге (метод Рунге-Кутты 4-го порядка):")
print("x\t\tu_h (h=0.05)\tu_H (h=0.1)\tОценка погр.")
print("-" * 70)
for i in range(len(x_coarse)):
    x = x_coarse[i]
    u_fine = y_rk4[2*i, 0]  # значение на мелкой сетке
    u_coarse = y_rk4_coarse[i, 0]
    error_estimate = runge_error_estimate(u_fine, u_coarse)
    print(f"{x:.2f}\t\t{u_fine:.6f}\t{u_coarse:.6f}\t{error_estimate:.2e}")
```

## Анализ результатов

```python
# Анализ максимальных погрешностей
max_error_euler = np.max(calculate_errors(y_euler[:, 0], y_exact))
max_error_rk4 = np.max(calculate_errors(y_rk4[:, 0], y_exact))
max_error_adams = np.max(calculate_errors(y_adams[:, 0], y_exact))

print("\nМАКСИМАЛЬНЫЕ ПОГРЕШНОСТИ:")
print(f"Метод Эйлера: {max_error_euler:.2e}")
print(f"Метод Рунге-Кутты 4-го порядка: {max_error_rk4:.2e}")
print(f"Метод Адамса 3-го порядка: {max_error_adams:.2e}")

# Проверка в конечной точке
u_final_exact = exact_solution(x_max)
u_final_rk4 = y_rk4[-1, 0]
final_error = abs(u_final_rk4 - u_final_exact)
print(f"\nЗначение в конечной точке x={x_max}:")
print(f"Точное решение: {u_final_exact:.8f}")
print(f"Рунге-Кутта: {u_final_rk4:.8f}")
print(f"Фактическая погрешность: {final_error:.2e}")
```

## Визуализация результатов

```python
# Графики
plt.figure(figsize=(15, 10))

# График 1: Сравнение решений
plt.subplot(2, 2, 1)
plt.plot(x_fine, y_euler[:, 0], 'b--', label='Метод Эйлера', linewidth=1, alpha=0.7)
plt.plot(x_fine, y_rk4[:, 0], 'g-', label='Рунге-Кутта 4', linewidth=2)
plt.plot(x_fine, y_adams[:, 0], 'r-.', label='Адамс 3', linewidth=1, alpha=0.7)
plt.plot(x_fine, y_exact, 'k-', label='Точное решение', linewidth=2, alpha=0.5)
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Сравнение численных методов')
plt.legend()
plt.grid(True, alpha=0.3)

# График 2: Погрешности
plt.subplot(2, 2, 2)
plt.semilogy(x_fine, calculate_errors(y_euler[:, 0], y_exact), 'b--', label='Погр. Эйлера', linewidth=1)
plt.semilogy(x_fine, calculate_errors(y_rk4[:, 0], y_exact), 'g-', label='Погр. Рунге-Кутта', linewidth=2)
plt.semilogy(x_fine, calculate_errors(y_adams[:, 0], y_exact), 'r-.', label='Погр. Адамса', linewidth=1)
plt.xlabel('x')
plt.ylabel('Абсолютная погрешность')
plt.title('Погрешности методов')
plt.legend()
plt.grid(True, alpha=0.3)

# График 3: Производная решения
plt.subplot(2, 2, 3)
plt.plot(x_fine, y_euler[:, 1], 'b--', label="Эйлер: u'(x)", linewidth=1, alpha=0.7)
plt.plot(x_fine, y_rk4[:, 1], 'g-', label="Рунге-Кутта: u'(x)", linewidth=2)
plt.plot(x_fine, y_adams[:, 1], 'r-.', label="Адамс: u'(x)", linewidth=1, alpha=0.7)
plt.xlabel('x')
plt.ylabel("u'(x)")
plt.title('Производная решения')
plt.legend()
plt.grid(True, alpha=0.3)

# График 4: Оценка по Рунге
plt.subplot(2, 2, 4)
runge_errors = []
for i in range(len(x_coarse)):
    error_estimate = runge_error_estimate(y_rk4[2*i, 0], y_rk4_coarse[i, 0])
    runge_errors.append(error_estimate)
    
plt.semilogy(x_coarse, runge_errors, 'mo-', label='Оценка по Рунге', linewidth=2)
plt.xlabel('x')
plt.ylabel('Оценка погрешности')
plt.title('Оценка точности по Рунге')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
