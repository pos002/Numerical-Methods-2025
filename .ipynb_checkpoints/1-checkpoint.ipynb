### Численное интегрирование

**Задача** (вариант $8$)

Требуется вычислить определенный интеграл

$$
I = \int_a^b f(x)  dx
$$

где $f(x) = \frac{1}{\cosh^2 x}$, $a = -1$, $b = 3$.

**Необходимо:**
1. Применить три численных метода: метод прямоугольников, метод трапеций и метод Симпсона.
2. Исследовать, как ошибка вычисления $|I_{\text{точн}} - I_{\text{прибл}}|$ зависит от шага сетки $h$.

### Теория

**Общая идея численного интегрирования**

Все три метода основаны на одной идее: разбить отрезок интегрирования $[a, b]$ на $n$ равных частей (с шагом $h = \frac{b-a}{n}$) и заменить сложную функцию $f(x)$ на более простую (константу, линейную или квадратичную функцию) на каждом маленьком отрезке $[x_{i-1}, x_i]$.

- **Узлы сетки:** $x_i = a + i \cdot h$, где $i = 0, 1, 2, ..., n$
- **Значения функции в узлах:** $y_i = f(x_i)$

**Метод прямоугольников (средних)**

На каждом отрезке $[x_{i-1}, x_i]$ функция аппроксимируется **константой**, равной значению функции в **середине** отрезка.

**Формула:**
$$
I_{\text{пр}} = h \cdot \sum_{i=0}^{n-1} f\left( \frac{x_i + x_{i+1}}{2} \right)
$$

или, что то же самое:
$$
I_{\text{пр}} = h \cdot \sum_{i=1}^{n} f\left( x_{i-1} + \frac{h}{2} \right)
$$

**Метод трапеций**

На каждом отрезке $[x_{i-1}, x_i]$ функция аппроксимируется **линейной функцией**, проходящей через точки $(x_{i-1}, y_{i-1})$ и $(x_i, y_i)$. Интеграл на таком отрезке – это площадь трапеции.

**Формула:**
$$
I_{\text{тр}} = \frac{h}{2} \cdot \left[ y_0 + y_n + 2 \sum_{i=1}^{n-1} y_i \right]
$$

$$
I_{\text{тр}} = h \cdot \left( \frac{f(a) + f(b)}{2} + \sum_{i=1}^{n-1} f(x_i) \right)
$$

**Метод Симпсона (парабол)**

На каждой паре отрезков $[x_{i-1}, x_i]$ и $[x_i, x_{i+1}]$ (всего на отрезке длины $2h$) функция аппроксимируется **квадратичным полиномом** (параболой), проходящим через три точки: $(x_{i-1}, y_{i-1}), (x_i, y_i), (x_{i+1}, y_{i+1})$. Метод требует четного количества отрезков $n$.

**Формула:**
$$
I_{\text{симп}} = \frac{h}{3} \cdot \left[ y_0 + y_n + 4 \sum_{\substack{i=1 \\ i \text{ нечетные}}}^{n-1} y_i + 2 \sum_{\substack{i=2 \\ i \text{ четные}}}^{n-2} y_i \right]
$$

(Суммирование по нечетным и четным индексам).

**Построение параболы в методе Симпсона**

*Идея построения параболы*

Мы берем **три последовательные точки** на графике функции:
- $(x_0, y_0)$
- $(x_1, y_1)$
- $(x_2, y_2)$

Где $x_1 = x_0 + h$, $x_2 = x_0 + 2h$

Через эти три точки можно провести **единственную параболу** (квадратичный полином):
$$
P(x) = Ax^2 + Bx + C
$$

**Вывод формулы для одного двойного отрезка**

Найдем интеграл от этой параболы на отрезке $[x_0, x_2]$:

$$
\int_{x_0}^{x_2} P(x)dx = \int_{x_0}^{x_2} (Ax^2 + Bx + C)dx = \left[ \frac{A}{3}x^3 + \frac{B}{2}x^2 + Cx \right]_{x_0}^{x_2}
$$

Подставляем пределы интегрирования. Учтем, что $x_2 = x_0 + 2h$:

$$
\int_{x_0}^{x_2} P(x)dx = \frac{A}{3}[(x_0+2h)^3 - x_0^3] + \frac{B}{2}[(x_0+2h)^2 - x_0^2] + C[(x_0+2h) - x_0]
$$

Раскрываем скобки:
- $(x_0+2h)^3 - x_0^3 = 6x_0^2h + 12x_0h^2 + 8h^3$
- $(x_0+2h)^2 - x_0^2 = 4x_0h + 4h^2$
- $(x_0+2h) - x_0 = 2h$

Подставляем:
$$
\int_{x_0}^{x_2} P(x)dx = \frac{A}{3}(6x_0^2h + 12x_0h^2 + 8h^3) + \frac{B}{2}(4x_0h + 4h^2) + C(2h)
$$

Теперь используем то, что парабола проходит через наши три точки:

1. $P(x_0) = Ax_0^2 + Bx_0 + C = y_0$
2. $P(x_1) = A(x_0+h)^2 + B(x_0+h) + C = y_1$
3. $P(x_2) = A(x_0+2h)^2 + B(x_0+2h) + C = y_2$

Решая эту систему уравнений, получаем удивительно простой результат:

$$
\int_{x_0}^{x_2} P(x)dx = \frac{h}{3}(y_0 + 4y_1 + y_2)
$$

**Откуда берется общая формула Симпсона**

Теперь представим, что у нас весь отрезок $[a,b]$ разбит на **четное число** $n$ отрезков. Мы можем применить формулу для параболы к каждой группе из трех точек:

- Для точек $x_0, x_1, x_2$: $\frac{h}{3}(y_0 + 4y_1 + y_2)$
- Для точек $x_2, x_3, x_4$: $\frac{h}{3}(y_2 + 4y_3 + y_4)$
- Для точек $x_4, x_5, x_6$: $\frac{h}{3}(y_4 + 4y_5 + y_6)$
- и т.д.

Складываем все эти интегралы:

$$
I = \frac{h}{3}[(y_0 + 4y_1 + y_2) + (y_2 + 4y_3 + y_4) + (y_4 + 4y_5 + y_6) + \cdots + (y_{n-2} + 4y_{n-1} + y_n)]
$$

Приводим подобные слагаемые:
- $y_0$ и $y_n$ встречаются по 1 разу
- Все **нечетные** $y_i$ ($y_1, y_3, y_5, \ldots$) встречаются с коэффициентом 4
- Все **четные** $y_i$ ($y_2, y_4, y_6, \ldots$, кроме $y_0$ и $y_n$) встречаются с коэффициентом 2

Получаем знаменитую формулу Симпсона:

$$
I_{\text{симп}} = \frac{h}{3} \left[ y_0 + y_n + 4\sum_{\text{нечетные } i} y_i + 2\sum_{\text{четные } i} y_i \right]
$$


**Исследование ошибки от шага сетки**

Теория утверждает, что ошибка численного интегрирования имеет следующий порядок малости относительно шага $h$:

- **Метод прямоугольников:** $R_{\text{пр}} \approx O(h^2)$
- **Метод трапеций:** $R_{\text{тр}} \approx O(h^2)$  
- **Метод Симпсона:** $R_{\text{симп}} \approx O(h^4)$

### Программная реализация

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# исходная функция
def f(x):
    return 1 / np.cosh(x)**2

a = -1  # нижний предел интегрирования
b = 3   # верхний предел интегрирования
```

```python
# точное значение интеграла (аналитическое решение)
def exact_integral(a, b):
    return np.tanh(b) - np.tanh(a)

I_exact = exact_integral(a, b)
print(f"Точное значение интеграла: {I_exact:.10f}")
```

```python
# метод прямоугольников (средних)
def rectangle_method(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)

    x_mid = (x[:-1] + x[1:]) / 2
    return h * np.sum(f(x_mid))
```

```python
# метод трапеций
def trapezoidal_method(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    return h * (0.5 * (y[0] + y[-1]) + np.sum(y[1:-1]))
```

```python
# метод Симпсона, требует четного количества отрезков
def simpson_method(f, a, b, n):
    if n % 2 != 0:
        n += 1  # делаем n четным
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)

    sum_odd = np.sum(y[1:-1:2])
    sum_even = np.sum(y[2:-1:2])

    return (h / 3) * (y[0] + y[-1] + 4 * sum_odd + 2 * sum_even)
```

```python
# различные количества разбиений
n_values = np.array([10, 20, 40, 80, 160, 320, 640])
h_values = (b - a) / n_values
```

```python
# массивы для хранения результатов
I_rect = np.zeros_like(n_values, dtype=float)
I_trap = np.zeros_like(n_values, dtype=float)
I_simp = np.zeros_like(n_values, dtype=float)

errors_rect = np.zeros_like(n_values, dtype=float)
errors_trap = np.zeros_like(n_values, dtype=float)
errors_simp = np.zeros_like(n_values, dtype=float)
```

```python
print("\nРезультаты вычислений:")
print("n\t h\t\t Rect Error\t Trap Error\t Simp Error")
print("-" * 65)

for i, n in enumerate(n_values):
    I_rect[i] = rectangle_method(f, a, b, n)
    I_trap[i] = trapezoidal_method(f, a, b, n)
    I_simp[i] = simpson_method(f, a, b, n)

    errors_rect[i] = abs(I_exact - I_rect[i])
    errors_trap[i] = abs(I_exact - I_trap[i])
    errors_simp[i] = abs(I_exact - I_simp[i])

    print(f"{n}\t {h_values[i]:.4f}\t {errors_rect[i]:.2e}\t {errors_trap[i]:.2e}\t {errors_simp[i]:.2e}")
```

```python
def calculate_order(h, errors):
    log_h = np.log(h)
    log_e = np.log(errors)

    k, C = np.polyfit(log_h, log_e, 1)
    return k

order_rect = calculate_order(h_values, errors_rect)
order_trap = calculate_order(h_values, errors_trap)
order_simp = calculate_order(h_values, errors_simp)

print(f"\nЭкспериментальные порядки сходимости:")
print(f"Метод прямоугольников: k = {order_rect:.4f} (теоретический: 2)")
print(f"Метод трапеций: k = {order_trap:.4f} (теоретический: 2)")
print(f"Метод Симпсона: k = {order_simp:.4f} (теоретический: 4)")
```

```python
# визуализация результатов

plt.figure(figsize=(15, 5))

# график 1: зависимость ошибки от шага
plt.subplot(1, 2, 1)
plt.loglog(h_values, errors_rect, 'o-', label='Прямоугольники', linewidth=2)
plt.loglog(h_values, errors_trap, 's-', label='Трапеции', linewidth=2)
plt.loglog(h_values, errors_simp, '^-', label='Симпсон', linewidth=2)

# теоретические линии для сравнения
h_fine = np.logspace(np.log10(h_values.min()), np.log10(h_values.max()), 100)
plt.loglog(h_fine, 0.1 * h_fine**2, 'k--', alpha=0.5, label='O(h²)')
plt.loglog(h_fine, 0.01 * h_fine**4, 'k:', alpha=0.5, label='O(h⁴)')

plt.xlabel('Шаг сетки h')
plt.ylabel('Абсолютная ошибка')
plt.title('Зависимость ошибки от шага сетки')
plt.legend()
plt.grid(True, alpha=0.3)

# график 2: сходимость методов
plt.subplot(1, 2, 2)
plt.semilogy(n_values, errors_rect, 'o-', label='Прямоугольники', linewidth=2)
plt.semilogy(n_values, errors_trap, 's-', label='Трапеции', linewidth=2)
plt.semilogy(n_values, errors_simp, '^-', label='Симпсон', linewidth=2)

plt.xlabel('Количество разбиений n')
plt.ylabel('Абсолютная ошибка')
plt.title('Сходимость методов интегрирования')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

$$
E_{\text{trap}} \approx -\frac{(b-a)}{12}\, h^{2}\, f''(\xi)

**Симпсон:**

E_{\text{simp}} \approx -\frac{(b-a)}{180}\, h^{4}\, f^{(4)}(\xi)

```python
import sympy as sp

x = sp.symbols('x')
f_expr = 1 / sp.cosh(x)**2

f2_expr = sp.diff(f_expr, x, 2)
f4_expr = sp.diff(f_expr, x, 4)

# превращаем в numpy-функции
f = sp.lambdify(x, f_expr, 'numpy')
f2 = sp.lambdify(x, f2_expr, 'numpy')
f4 = sp.lambdify(x, f4_expr, 'numpy')
```

```python
xs = np.linspace(a, b, 5000)
max_f2 = np.max(np.abs(f2(xs)))
max_f4 = np.max(np.abs(f4(xs)))

errors_rect_theory = (b - a) * h_values**2 * max_f2 / 24
errors_trap_theory = (b - a) * h_values**2 * max_f2 / 12
errors_simp_theory = (b - a) * h_values**4 * max_f4 / 180
```

```python
print("\nМетод прямоугольников:")
print("n\t h\t\t Числ. ошибка\t Теор. оценка")
for i in range(len(n_values)):
    print(f"{n_values[i]}\t {h_values[i]:.4f}\t {errors_rect[i]:.2e}\t {errors_rect_theory[i]:.2e}")

print("\nМетод трапеций:")
print("n\t h\t\t Числ. ошибка\t Теор. оценка")
for i in range(len(n_values)):
    print(f"{n_values[i]}\t {h_values[i]:.4f}\t {errors_trap[i]:.2e}\t {errors_trap_theory[i]:.2e}")

print("\nМетод Симпсона:")
print("n\t h\t\t Числ. ошибка\t Теор. оценка")
for i in range(len(n_values)):
    print(f"{n_values[i]}\t {h_values[i]:.4f}\t {errors_simp[i]:.2e}\t {errors_simp_theory[i]:.2e}")
```
